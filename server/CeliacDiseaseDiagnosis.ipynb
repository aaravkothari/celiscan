```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
celiac_df = pd.read_csv("celiac_disease_lab_data.csv")
celiac_df.iloc[-15:]
```

```python
celiac_df['Diarrhoea'].value_counts()
```

```python
celiac_df.info()
```

```python
celiac_df.describe()
```

```python
celiac_df = celiac_df.drop(['Diabetes Type', 'Marsh', 'cd_type', 'Gender'], axis=1)
celiac_df = celiac_df.dropna()
celiac_df
```

```python
X = celiac_df.drop(['Disease_Diagnose'], axis=1)
y = celiac_df['Disease_Diagnose']
```

```python
numerical = X.drop(['Diabetes', 'Diarrhoea', 'Abdominal', 'Short_Stature', 'Sticky_Stool', 'Weight_loss'], axis = 1)
numerical.head()
```

```python
categorical = X.filter(['Gender', 'Diabetes','Diarrhoea', 'Abdominal', 'Short_Stature', 'Sticky_Stool', 'Weight_loss'])
categorical.head()
```

```python
cat_numerical = pd.get_dummies(categorical,drop_first=True)
cat_numerical.head()
```

```python
X = pd.concat([numerical, cat_numerical], axis = 1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform (X_test)
```

```python
from sklearn import svm

# Initialize the Support Vector Classifier
svm_clf = svm.SVC(probability=True)

# Fit the classifier to the training data
classifier = svm_clf.fit(X_train, y_train)

# Predict on the test data
y_pred = classifier.predict(X_test)

# Scaling a single record
single_record = sc.transform(X.iloc[900].to_frame().T)

# Making predictions on the single record
predicted_class = classifier.predict(single_record)

# Print the predicted bankrupt class outcome
print("Index [900] Single Value Prediction Comparison:\n")
print(f"Actual Class: 1 | Predicted Class: {predicted_class[0]}\n\n")

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Print the confusion matrix
print("Confusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))

print("\n")

# Print the classification report
print(classification_report(y_test, y_pred))

# Print the accuracy score
print("Accuracy:", accuracy_score(y_test, y_pred))

from sklearn.model_selection import cross_val_score

# Print the cross-validation score
print("\n\nAccuracy K-Fold Cross Val:\n")
print(cross_val_score(classifier, X, y, cv=5, scoring="accuracy"))
```
